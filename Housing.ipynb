```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df.shape
```

```python
df1=pd.read_csv("test.csv")
df1.head()
df1.shape
```

```python
main = pd.concat([df,df1],sort=False)
```

```python
def clean(column,main):
    shape = main.shape
    type1 = main.dtypes
    null1 = main.isnull().sum()
    column1 = main.columns
    mean=main[main[column]!=np.nan][column].astype(float).mean()
    main[column]=main[column].replace(np.nan,str(mean)).astype(float)
    return main
```

```python
clean("LotFrontage",main)
clean("MasVnrArea",main)
clean("GarageYrBlt",main)
clean("GarageCars",main)
clean("GarageArea",main)
clean("BsmtFinSF1",main)          
clean("BsmtFinSF2",main)          
clean("BsmtUnfSF",main)
clean("BsmtFullBath",main)        
clean("BsmtHalfBath",main)  
clean("TotalBsmtSF",main)
```

```python
main.isnull().sum()
```

```python
main = main.drop(["Alley","MasVnrType","FireplaceQu","GarageType","GarageFinish","GarageQual","GarageCond","PoolQC","Fence","MiscFeature",
                 "BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2","Electrical","Exterior1st","Exterior2nd","SaleType",
                 "Utilities","KitchenQual","Functional","MSZoning"],1)
```

```python
numdf = main.select_dtypes(include=np.number)
categorical = main.select_dtypes(exclude=np.number)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
cols=list(numdf) 
numdf=scaler.fit_transform(numdf)
numdf=pd.DataFrame(numdf,columns=cols)
numdf.head()
```

```python

categorical = pd.get_dummies(categorical)
categorical.head()
```

```python
numdf.reset_index(drop=True, inplace=True)
categorical.reset_index(drop=True, inplace=True)
main_df=pd.concat([numdf,categorical],1)
main_df.head()
main_df.shape
```

```python
X=main_df.drop(["SalePrice"],1)
y=main_df["SalePrice"]
```

```python
X_train = X[:1460]
X_test = X[1461:]
y_train = y[:1460]
y_test = y[1461:]
```

```python
X_test = X_test.fillna(X_test.mean())
```

```python
from sklearn import metrics
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
```

```python
lin_reg.fit(X_train,y_train)
y_pred= lin_reg.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
Regressor = KNeighborsRegressor(n_neighbors=5)
Regressor.fit(X_train,y_train)
y_pred = Regressor.predict(X_test)
```
